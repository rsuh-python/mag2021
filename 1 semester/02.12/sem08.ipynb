## Задание 5.

Напишем функцию для красивого вывода словаря. Функция должна выводить ключи и значения в виде:

key :: value

Для продвинутых: попробуйте выравнивать ключи и значения по самому широкому полю (соответственно ключа и значения). 

```python
def printdict(d):
    lenkeys = max([len(str(x)) for x in d.keys()]) + 2
    lenvalues = max([len(str(x)) for x in d.values()]) + 2
    for k, v in d.items(): # [(k, v), (k, v)...]
        print(f'{k:<{lenkeys}}::{v:>{lenvalues}}') # f'{var:<{lenkeys}}::
        
printdict(months)
```
