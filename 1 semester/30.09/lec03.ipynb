В начале занятия доразобрали методы строк: split() и join(). 

str.split() делит строчку на куски. По умолчанию границами он считает пробельные символы (то есть, собственно пробел, табуляцию, перенос на новую строку...). Таким образом, если не указать в скобках границу для разделения, сплит работает так:

```python
'this is     my\t\tspacious   string\nand another   one'.split()
```

Метод split() возвращает объект класса **список**. Можно и явно задать этому методу знак, который считать границей для разделения:

```python
'this!is!my!!string!'.split('!')
```

Отличительная особенность split() без аргументов состоит в том, что он съедает все пробелы и не создает пустых строк. Когда мы передаем аргумент явно, он считает, что между !! есть некое пустое пространство, которое и выделяет в отдельную пустую строку ''. 

Метод join() работает зеркально противоположно: он принимает в качестве аргумента итерируемый объект (строку, список) и между его элементами вставляет строку, к которой применяется. 

    list of strings = wholestring.split(splitter)
    wholestring = splitter.join(list of strings)


Также мы вкратце обговорили множественное присваивание в питоне. 

Питон умеет присваивать сразу несколько объектов в несколько переменных:

    x, y = 1, 2

В этом случае он положит единицу в x, а двойку в y. Можно пользоваться множественным присваиванием, чтобы поменять значения переменных местами:

    x, y = y, x
    
Тогда значение x окажется в y и наоборот. 
Наконец, можно присваивать в несколько переменных один список, если у этого списка столько же элементов, сколько переменных. Метод split() возвращает список, если мы заранее знаем, сколько должно быть строк, то можно так:

    x, y = input().split()
    
Если введена строка с одним пробелом, то в x окажется первая часть, а в y - вторая.

Для любопытных: можно складывать неограниченное количество объектов в переменную со звездочкой (\*lst, last_word = input().split()), но при этом должна быть еще какая-то переменная. Штуки со звездочками, я надеюсь, мы успеем рассмотреть, когда будем обсуждать функции.

## Форматирование строк

Форматирование строк используется обычно для "красивишного" вывода, для того, чтобы выводить таблички. Основная идея состоит в том, что у нас есть какая-то строка-шаблон, в которую мы подставляем наши переменные (их еще называют джокерами): например, у нас есть переменные "имя" и "фамилия", которые меняются, а нам нужно много раз выводить в печать "Имя: \<имя\> Фамилия: \<фамилия\>". Методы форматирования ровно это и позволяют делать: вставляют переменные в строку-шаблон без нужды использовать конкатенацию или унылое перечисление. 

Сравните:

```python
name, surname = input().split()
print('Имя:', name, '\nФамилия:', surname)
print('Имя: ' + name + '\nФамилия: ' + surname)  # конкатенация строк, ей часто пользуются не очень хитрые программисты
print(f'Имя: {name}\nФамилия: {surname}')
```

Способов форматирования в 3-м питоне есть два:

1. Метод строк str.format()
2. f-strings

str.format() применяется к строке-шаблону, а переменные, которые нужно вставить внутрь шаблона, передаются аргументами. В самом шаблоне ставятся фигурные скобочки в местах, куда нужно их вставить, а внутри скобочек еще можно расставить нумерацию.

```python
x = 2.345678
y = 1234.23456
print('x: {}, y: {}'.format(x, y))
print('y: {1}, x: {0}'.format(x, y))  # помним, что нумерация идет с нуля
```

f-строки работают еще проще: мы просто вписываем названия переменных внутрь фигурных скобок. 

Два примечания:
1. Чтобы вывести фигурную скобку в печать, нужно ее удвоить: {{}} выведет сами фигурные скобки. 
2. Чтобы использовать объект с кавычками внутри фигурных скобок, нужно использовать разные варианты кавычек: f"{'string'}"

У f-строк, как и у метода format, есть не только возможность подставлять переменные в шаблон, но и выводить содержимое этих переменных с какими-то параметрами. 

Мы рассмотрели два параметра:

1. Выравнивание (<^>) и паддинги (сколько пространства оставлять под наш объект)
2. Округление и формат чисел с плавающей точкой (notation)

Все параметры передаются после двоеточия внутри фигурных скобок. Пробелы там ставить нельзя, потому что там каждый символ значащий! Питон будет пытаться расценить лишний пробел как параметр. 

```python
print(f'{x:<15}|{y:>15}')  #отведет для поля "х" 15 мест и поставит само значение слева, а для поля "y" - справа. 
```

Также в параметрах можно передавать, до какого знака после запятой округлить наше число и в каком формате его представить. Форматов знаем два: экспоненциальный и обычный с плавающей точкой (есть еще с фиксированной точкой). Экспоненциальное представление - это когда к числу добавляется буква е (экспонента) с какой-то степенью. Нам, к счастью, с таким форматом обычно не приходится работать. Однако питон, если ему явно не указать формат f, выбирает тот, который ему больше нравится, и иногда это может оказаться формат с е. 

```python
print(f'{x:<15.2f}|{y:>15.3e}')
```

Таким образом, что может быть в параметрах у форматированной строки:

    {<сам объект>:<alignment><padding><round up><notation>}

## Циклы

В парадигме структурного программирования рассматривается три понятия:
- последовательность
- ветвление
- цикл

Про первые два мы уже поговорили, теперь поговорим о третьем. 

Цикл - это некоторый набор команд, которые нужно повторить несколько раз (одно повторение = итерация). Циклы бывают двух видов:

1. Когда мы заранее знаем, сколько раз нужно повторить. 
2. Когда мы должны повторять, пока выполняется какое-то условие. 

Цикл for - это первый случай: мы заранее знаем, сколько раз повториться. Тело цикла в общем виде выглядит так:

    for <переменная цикла> in <итерируемый объект>:
        <команда1>
        <команда2>
    else:
        <команда3>
        
else, как и для if, здесь не обязательный. Что здесь что?

- переменная цикла - заранее не определяется; в эту переменную на каждой новой итерации подставляется следующий элемент из итерируемого объекта. Например, если итерируемый объект - строка, то в переменную цикла подставляются символы по очереди. Эта переменная не исчезает, когда отработал цикл, но обращаться к ней лучше не стоит. 
- итерируемый объект - может быть строкой, списком, итератором, множеством или словарем, главное - это чтобы он состоял из нескольких элементов, которые можно перебрать. 
- команды цикла выполняются в теле цикла (точно так же, как в if) и выделяются индентингом. 

В связи с for неплохо обсудить функцию range(). Эта функция создает диапазон чисел (ее аргументы работают по такому же принципу, как в срезе: start, stop, step), по которому можно итерироваться. 

```python
s = 'qwerty'
for char in s:
    print(char, end='\t')  # здесь я специально прописываю end, потому что на каждой итерации print будет выполняться по-новой и переносить на новую строку, а я этого не хочу
```

```python
s = 'qwerty'
for i in range(len(s)):
    print(s[i], end='\t')
```

```python
for i in range(10):
    print(i, end=' ')
print()
for j in range(1, 11):
    print(j, end=' ')
print()
for k in range(1, 11, 2):
    print(k, end=' ')
```

Вторая разновидность цикла - цикл while. Он больше похож на условные инструкции:

    while <condition>:
        <command1>
        <command2>
    else:
        <command3>
        
else опять необязательный. По правде говоря, с циклами else довольно редко используется: но есть случаи, когда он может пригодиться. 

Цикл while выполняется, пока выполняется условие, и проверяет это условие каждый раз, когда начинает новую итерацию. Если условие перестает выполняться, он один раз выполняет else (если тот есть). 

```python
s = input()
while s:  # если s - пустая строка, то bool(s) = False; not False = True
    print(s)
    s = input()
```

В связи с циклами нужны две команды: break и continue. 

- break - когда программа встречает эту команду, она полностью выходит из цикла, наплевав на все инструкции, которые шли за break. 
- continue - когда программа встречает эту команду, она плюет на все инструкции, которые идут после нее, и переходит на новую итерацию, но из цикла не выходит. 

```python
while True:
    a = input()
    if not a:
        print('Введена пустая строка')
        break
    print(a)
```

```python
for i in range(3):
    x = int(input())
    if not x:
        break
else:
    print('0 не вводился')
```

Примерно в таких ситуациях и нужен else: если выполняется break, не выполнится else. 

```python
for i in range(1, 11):
    if i % 2:
        continue
    print(f'{i} - четное число')
    print(i ** 2)
```

Напоследок: циклы и условные инструкции могут вкладываться друг в друга, и это часто бывает. 

## Сложность алгоритмов

Обычно изучающим питон про это особенно не рассказывают, но это важная часть программистской жизни. Я считаю, что компьютерные лингвисты тоже должны об этом знать: наши инструменты могут быть очень медленными (да и питон небыстрый), поэтому иногда критично, выполняется ли наша программа за три дня или за три месяца. Код может работать месяцами без перерывов, имейте это в виду (обычно так долго работают всякие нейронки). 

Любая программа - это алгоритм, воплощенный в командах для машины. Алгоритмы имеют свою сложность выполнения: по-простому, это t = f(n): зависимость времени выполнения от объема обрабатываемых данных. Стандартная нотация - O(...). 

- Константное время: O(const). Это означает, что сколь бы мы ни скормили данных нашей программе, она всегда будет отрабатывать примерно за одно и то же время. Отлично! Жаль, таких программ почти не бывает. 
- Линейное время: O(n). Это означает, что мы скормили программе n данных (3 мегабайта, например), а она их обрабатывает за a * n + b времени, то есть, если 3 мегабайта обработает за 3 секунды, то 9 мегабайт - за 9 секунд. 
- Квадратичное время: O(n ** 2). А вот это уже плохо: это означает, что если 3 мегабайта программа обрабатывает за 3 секунды, то 6 мегабайт - уже за 9 секунд, а 12 мегабайт - за 81 секунду. 

На этом, конечно, варианты не исчерпываются, например, алгоритм сортировки (лучший) работает со сложностью O(n \* log(n)). А худший вариант - это O(exp ** n), но про это даже думать не хочется. 

Вложенные циклы часто означают квадратичное время (но не всегда). Поэтому предпочтительно их избегать - но иногда без них все равно не обойтись. 

## Списки и изменяемые объекты

В основном списки будем рассматривать на следующем занятии, пока успели только обсудить самое начало. 

Списки - это сложные, составные объекты. Списки обозначаются в квадратных скобочках:

    [1, 2, 3] - список чисел
    ['1', '2', '3'] - список строк
    
Внутри списка могут находиться совершенно любые объекты. Необязательно даже одного типа. Могут быть списки списков, списки списков списков, списки списков... ну вы поняли. Кстати, такие вот списки списков списков... используются для работы с матрицами и прочими многомерными жуткостями. Ну, или могут использоваться (с матрицами в питоне работает библиотека numpy, до которой мы с вами, надеюсь, когда-нибудь доживем).

Списки - итерируемые изменяемые объекты. 

Изменяемые означает то, что они могут изменяться по частям. 

Списки тоже индексируются, а значит, к спискам можно применять все те же срезы, точно так же, как к строкам. С единственной разницей: один элемент строки изменить нельзя, а в списке пожалуйста. 

```python
lst = [1, 2, 3]
lst[0] = 5
lst
```

С изменяемостью списков связана основная сложность питона, которую будем обсуждать на следующем занятии. Пока же - как завести список?

- явно задать в коде:

        lst = [1, 2, 3]
        
- превратить из другого итерируемого объекта:

        lst = list(range(10))
        lst = list('qwerty')
        
- использовать list comprehension + генератор:

        lst = [func(x) for x in <iterable> if <cond>]
        
- завести пустой список и понадобавлять туда элементов методом append():

        lst = []
        for char in s:
            lst.append(char)
            

Самое интересное из этого всего - представление списка (list comprehension). Чем отличается от генератора?

Генератор - такой особый объект в питоне; он итерируемый, но в момент, когда вы его создаете, он не вычисляется. Что это значит? если у нас в генераторе написано 

    (x ** 2 for x in lst),
    
он не кинется сразу же вычислять все квадраты для всех иксов в списке lst, а будет их выдавать по одному, если его попросить. Генератор - птица гордая, пока не пнешь, не полетит! (зато летает быстро)

list comprehension - это что-то вроде list(), только в квадратных скобочках: все, что стоит внутри квадратных скобочек, превращается в список. Просто если это какие-то готовые объекты, то там особенно нечему превращаться, а если внутри квадратных скобок стоит генератор, скобки заставят его выдать все его элементы и превратят их в список. 

```python
A = list(range(1, 11))
B = [x ** 2 for x in A]
C = [x for x in A if x % 2]
print(B, C)
```

Взглянем на сравнение скорости работы генератора и цикла из семинара:

```python
string = 'abc' * 1000
%timeit [char for char in string if char != 'a']
```

```python
%%timeit
L = []
for char in string:
    if char != 'a':
        L.append(char)
```

Заодно познакомимся с чудесными магическими функциями IPython, которых нет в CPython (там есть свои методы, не настолько удобные). Функция %timeit заставляет питон гонять команду, которая идет после этой функции, тысяч десять раз и оценивает, с какой в среднем скоростью выполняется эта команда. То же, но с двумя %, гоняет всю ячейку. 

Цифры на каждом новом запуске будут меняться (потому что скорость работы зависит от того, насколько в конкретную секунду загружен процессор, и все такое прочее), но одно будет совершенно точно: генератор отрабатывает быстрее цикла, иногда почти в два раза. 

Как, кстати, читать эти обозначения, которые выводит timeit: µs - микросекунды, ns - наносекунды. То есть, среднее значение скорости равно скольки-то микросекундам с отклонением в сколько-то наносекунд (или микросекунд) на прогон (loop): а прогнано было 10000 раз за 7 циклов (число прогонов и циклов может меняться). Это означает, что питон погонял нашу строчку кода 7 * 10000 раз.
