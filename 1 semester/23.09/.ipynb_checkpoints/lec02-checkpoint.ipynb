В начале занятия мы с вами вспоминали и доразбирали, что такое функции в питоне. 

Функция - это точно такой же *объект*. Функции на самом деле - заранее скомпилированные кусочки кода; например, если нам постоянно нужно выполнять один и тот же набор действий в разных местах программы, мы можем копировать команды, а можем их вынести из программы и оформить в функцию: функция будет вызываться из разных частей программы и выполнять этот самый набор действий. Подробнее будем говорить в теме про функции. 

Таким образом, функция - что-то вроде заводика, логическое устройство, которое выполняет какие-то операции. Функция может принимать что-то на вход (и обрабатывать это), может ничего не принимать; точно так же функция может возвращать какой-то объект - результат своей работы - а может ничего не возвращать. 

Функция input() обычно ничего не принимает (у нее нет аргументов, но можно передать ей строчку: тогда она ее выведет, прежде чем выполнить свою работу: считать данные, вводимые пользователем), но всегда возвращает строчку (не забывайте, что input() возвращает ТОЛЬКО строчку, а значит, если нужно считать число, придется явно преобразовать эту строчку в число). 

Функция print() обычно что-нибудь да принимает (то, что нужно вывести в печать), но не возвращает ничего: она выполняет свои действия - выводит принятые объекты на экран - только и всего. 

Еще мы повторили, что такое операторы в питоне. 

Операторы - это любые символы, которые производят операции с объектами и (обычно) возвращают новые объекты. К операторам относятся арифметические (+, -...) и логические. "=" - тоже оператор! (Присваивания)

## Тип bool и условные инструкции

Тип данных bool - особый вид объектов, разновидность int (в некот. языках программирования такого типа просто нет, и программисты в них обходятся int). Это логическое значение "истинно" и "ложно". У bool, таким образом, есть только два объекта: True и False (обязательно с заглавной буквы!). Bool - неизменяемый, как и int. 

Любой объект можно превратить в bool явным образом:

       a = 10
       bool(a)
       
Если мы превращаем в bool число, то все, кроме 0, превратится в True, а 0 - в False. Если мы превращаем в bool строку, то любая строка, кроме пустой строки '', превратится в True (обычно "пустые" объекты - False, а непустые - True). Этим свойством очень часто пользуются. 

Кроме арифметических операторов, в питоне существуют и логические. Все числа можно сравнивать между собой!

- \> - строго больше
- \< - строго меньше
- \>= - больше или равно
- \<= - меньше или равно
- == - равно (**не путайте с = !**)
- != - не равно

В питоне можно составлять выражения с логическими операторами:

        3 < 5 > 2 ...
        
Такие выражения *вычисляются*, и в результате получается объект типа bool. Вычисляются они обычно слева направо, можно нанизывать несколько таких операторов. 

Помимо символьных операторов, есть еще операторы:

- and
- or
- not

Они обозначают, соответственно, логическое и, логическое или и логическое не и используются для того, чтобы объединять объекты типа bool:

        True and False = False
        True or False = True
        not True = False
  
Логическое и работает как умножение (1 * 0 = 0) и имеет высший приоритет, чем логическое или (1 + 0 = 1). Поэтому в таких выражениях при надобности имеет смысл использовать скобки:

        (True or False) and True = True

Потом мы с вами немного вернулись в прошлое и вспомнили про парадигмы программирования. Парадигма - это система философских взглядов; так вот, в прошлом веке, когда только появились ЯП сложнее ассемблера, люди придумали оператор go to. То есть, программа выполняется обычно линейно, команда за командой; оператор go to заставляет компьютер при выполнении программы переходить на строку, указанную программистом. Это сильно запутывает программы, потому что go to позволяет перейти вообще на любую строчку. 

Писать такие программы, чтобы их можно было не только выполнять, но и *читать*, очень важно: код обычно бывает нужно *поддерживать* (то есть, постоянно дорабатывать), к тому же, над большими проектами обычно работает много людей. Конвенции придуманы именно для того, чтобы код было удобно читать человеку! Даже есть такая метрика - сложность кода, которая про то, насколько ваш код легко читать человеку. 

Программы, в которых оператор go to постоянно заставлял прыгать со строчки на строчку, было очень сложно читать, и ученый инженер-программист Дейкстра (https://ru.wikipedia.org/wiki/%D0%94%D0%B5%D0%B9%D0%BA%D1%81%D1%82%D1%80%D0%B0,_%D0%AD%D0%B4%D1%81%D0%B3%D0%B5%D1%80_%D0%92%D0%B8%D0%B1%D0%B5) рассердился и написал очень известную статью, в которой предложил навсегда отказаться от этого оператора и структурировать свои программы. 

Так появился структурный подход (на самом деле все парадигмы - более или менее про то, как писать код так, чтобы там все было понятно). Структурное программирование предполагает, что, прежде чем написать любую программу, человек придумывает алгоритм и рисует его блок-схему (https://ru.wikipedia.org/wiki/%D0%91%D0%BB%D0%BE%D0%BA-%D1%81%D1%85%D0%B5%D0%BC%D0%B0). Программа уже потом пишется в соответствии с этой блок-схемой. 

Примерно все программы, которые мы с вами будем писать в первом семестре, можно выполнять в соответствии с заветами Дейкстры. :) Кстати, когда решаете какую-то задачку, вообще бывает полезно порисовать на бумажке. С хитрыми задачками на логику очень помогает, я всегда рисовала в своем первом семестре. 

В структурном подходе есть три главных понятия:

- последовательность
- ветвление
- цикл

Последовательное выполнение - это когда наша программа выполняет команду за командой в той последовательности, в какой они написаны (если в вашей IDE, кстати, не включена нумерация строк в программе, включите: очень удобно). 

Ветвление - это условные инструкции: когда у нас есть какое-то условие, выполнение которого отправляет программу под одной ветке, а невыполнение - по другой (веток может быть много, в других ЯП есть такая штука, как switch case: для большого количества веточек; в питоне версии 3.10 тоже хотят ее имплементировать). 

О циклах поговорим на следующем занятии.

В питоне условные инструкции - это if elif else.

шаблон такой инструкции выглядит так:

        if <condition1>:
            <command1>
            <command2>
        elif <condition2>:
            <command3>
            <command4>
        ...
        else:
            <command5>
            
Здесь мы знакомимся с понятием *индентинга*: четыре пробела, которыми отбивается каждая команда внутри веточки инструкции, называются "индентинг". Этими четырьмя пробелами выделяется тело инструкции, и так становится наглядно видно, что у нас всего 3 веточки (многоточие стоит потому, что elif может повторяться несколько раз), и когда программа доходит до этой инструкции, она выполняет только команды в одной из них. 

Обязательным в инструкции if elif else является только if: остальные веточки при ненадобности можно не писать. Таким образом, часто можно встретить либо одинокий if, либо if-else; if-elif возможно, но обычно в таком случае удобнее if-else.

Кстати, программисты на других языках часто ненавидят питон за его индентинг. :) потому что в языках типа С тело инструкции выделяется явно {}, а индентинг там тоже принято делать, но он нужен только для читаемости кода. В целом это стандартный для питона подход: питонисты стремятся минимизировать формальные элементы в коде и поэтому избавились от {}, а заодно и от многого другого (необходимости явно задавать тип данных, например...)

## Строки

Один из самых полезных типов данных для лингвиста - это, конечно, строчка. Строка в питоне всегда пишется в кавычках, унарных или двойных:

        'qwerty'
        "qwerty"
        
Какие вы используете кавычки, неважно, главное, чтобы они были парные. Если мы внутри строки хотим использовать апостроф, можно его заэскейпить: 'king\\'s crown'. 

Если вам нужна строчка, в которой содержится несколько подстрочек (multiline), можно использовать тройные кавычки:

        '''это очень длинная строка
        в ней несколько подстрок
        каждая отделена переносом на новую строчку "\n"
        '''
        
Такие тройные кавычки позволяют внутри использовать одинарные кавычки и двойные, а также часто используются для комментирования кода. 

Комментарии - это вообще такая важная вещь. Коротенькие комменты можно оформлять с помощью #. # ставится один раз, и все от # до конца строчки считается комментарием и игнорируется интерпретатором питона. Комментарии, ясное дело, нужны только другим программистам, которые будут читать ваш код (или вам самим). Привычка писать комменты - хорошая, главное найти золотую середину и не комментить каждый пшик. 

Тройные кавычки """ """ используются для того, чтобы создавать docstrings - это такие комментарии, которые пишутся в определении функции (мы с вами научимся это все делать) и выводятся функцией help. 

Строчки - это неизменяемые объекты, как и числа. 

Что значит *неизменяемый* объект? Неизменяемость - это такой признак, который связан с главной особенностью питона. Питон автоматически управляет памятью, сам размещает свои данные в оперативе, и для этого ему нужно разделять объекты на изменяемые и неизменяемые. Понятнее это станет, когда изучим все встроенные типы данных. 

Неизменяемость означает, что наш объект нельзя изменить по частям, можно только перезаписать. 

        а = 5
        а += 1
       
Первая инструкция здесь складывает в переменную а ссылку, которая указывает, где в оперативной памяти лежит объект "5". Вторая инструкция произведет операцию инкрементации: увеличит 5 на 1 и вернет **новый** объект: 6. Если бы у нас был просто +, то пришлось бы написать b = a + 1, чтобы не потерять этот новый объект (потому что объекты, на которые не указывает ни одна переменная, скушиваются сборщиком мусора питона, чтоб место не занимали). Оператор += говорит питону "вычисли новый объект и сохрани его ссылку по прежнему адресу". Значит, теперь в переменной а будет храниться ссылка на новый объект "6", а на объект "5" больше не будет указывать ни одна ссылка, и он потеряется. 

Пока что мы имели дело только с простыми объектами: числами. Числа - элементарные объекты, они не разложимы по кусочкам (точнее говоря, человек может вообразить, что 123 - это объект, который можно разложить на 1, 2 и 3, но питон этого не понимает). 

Строка - неэлементарный объект: в строке есть отдельные символы. Мы можем обращаться к каждому такому символу по отдельности, но не можем их изменять внутри строки. 

Кстати, если это вам еще не пришло в голову: одиночный символ для питона - тоже строка. Этим питон отличается от С-подобных языков, где char и массив символов - разные типы данных. 

Как же нам обращаться к строке по частям? Для этого у строк есть индексация: каждый символ в строке пронумерован (от 0: в питоне практически все всегда считается с 0). Соответственно, к i-му символу строки можно обратиться с помощью специального синтаксиса:

        s = 'qwerty'
        s[0] - 'q'
        
(я специально написала -, а не =, потому что присвоить в s\[0\] мы ничего не можем). 

Индексация позволяет нам работать не только с отдельными символами строчки (например, перебирать их: тот факт, что мы можем перебрать строчку по символам, делает ее *итерируемым* объектом), но и с частями строки. Часть строки обычно называют "срез". 

Срезы берутся тоже с помощью квадратных скобок, а еще с помощью :. 

Общий вид среза: 

        str[start:stop:step]
        
Start - это первый индекс, с которого мы начинаем брать срез. Stop - индекс конца среза. Имейте в виду, что все интервалы в питоне такие: \[...). 

Это означает, что левая граница включена в интервал, а правая - нет. То есть, если мы пишем в качестве конца интервала 4, элемент с индексом 4 уже не войдет, а крайним окажется 3. Почему это так сделано? А потому, что считаем с нуля: тогда, если мы пишем "взять до 4 символа", получится, что мы и берем 4 символа, хотя 4-й символ, если считать с нуля - это на самом деле 5-й. 

Наконец, шаг - это берем ли мы все символы подряд, через один или даже задом наперед. 

```python
s = 'qwerty'
print(s[1])
print(s[1:4])
print(s[1:6:2])
print(s[:4])
print(s[1::2])
print(s[:6:2])
print(s[-1])
print(s[2:-1])
```

В ячейке выше можете погонять код, чтобы разобраться, как работают срезы. Одно из преимуществ индексации - она закольцована; таким образом, -1 символ - это 0 - 1, то есть, мы сдвигаемся с первого символа влево и попадаем в конец строки. Поэтому -1 символ - это последний символ в строке. Это удобно, когда мы не знаем, сколько всего символов в строчке, или они меняются (например, пользователь вводит строку). 

Любой из трех параметров среза можно опускать. Если он не крайний, то просто ставим двоеточие, а если крайний, то двоеточие опускаем. 

Наконец мы обсудили, что такое методы, и рассмотрели несколько методов строк. 

### Разница между методами и функциями

На самом деле методы - это просто разновидность функций: это такие же функции, но их особенность в том, что они *приписаны* к какому-то конкретному типу данных и могут вызываться только с его объектами. Это отражено в синтаксисе: обычно методы вызываются через точку от объекта, к которому они применяются. 

На сегодня вам достаточно запомнить одно:

        string = 'qwerty'
        len(string) - функция вызывается сама по себе, она независимая. Может в качестве аргумента принимать что хочет. 
        string.count('q') - метод вызывается с объектом, к которому он применяется, через точку. 
        
Забегая вперед, для любопытных и более продвинутых:

Главное внутреннее различие между функцией и методом - в месте, которое они занимают в иерархии объектов питона. На вершине этой иерархии находится модуль (\_\_main\_\_). Функции, определенные в этом модуле, принадлежат к нему; наравне с функциями в модуле могут определяться классы. Методы находятся уровнем ниже: они определяются *внутри* классов. Различие в синтаксисе вызвано просто обычным стремлением питон-программистов сократить все на свете. На самом деле стандартный синтаксис вызова функций и методов почти идентичен, это хорошо видно на функциях, которые мы импортируем из других модулей:

        import math
        math.sqrt() - функция, принадлежащая модулю math
        
        import pandas
        pandas.Dataframe.from_dict() - в модуле pandas есть класс Dataframe, у которого есть метод класса from_dict
        
Просто функции, которые определяются в главном модуле (нашем скрипте) или являются встроенными, можно писать без обозначения их принадлежности, а для методов классов на самом деле работает упрощенный синтаксис:

        str.count(string, 'q') - так должно быть по логике
        string.count('q') - упрощенный синтаксис, который питонисты придумали, чтобы упростить себе жизнь и усложнить ее всем остальным
        


Итак, функция, которая чаще всего применяется к строчкам:

```python
s = 'qwerty'
len(s)
```

Все методы строк не изменяют сами строки (строки неизменяемые!), а возвращают какие-то новые объекты.

Методы (и оператор) поиска по строке:

(возвращают в случаях 1-3 - int, в случае 4 - bool)

1. s.find(<что ищем>) - возвращает -1, если ничего не нашел, или индекс первого встретившегося элемента
2. s.rfind(<что ищем>) - то же самое, только ищет с правого конца. 
3. s.count(<что считаем>) - считает, сколько раз в строке s встретился элемент. 
4. '...' in s - оператор, который проверяет, есть ли элемент '...' в строке s.

Методы работы с регистром (строчные и заглавные буквы):

(возвращают новую строку)

1. s.upper() - делает все буквы в строке s заглавными
2. s.lower() - делает все буквы в строке s строчными
3. s.capitalize() - делает первую букву в строке s заглавной
4. s.title() - делает все первые буквы слов в строке (если строка из слов, разделенных пробелами) заглавными

```python
a = 'qwerty'
A = 'QWERTY'
t = 'lord of the rings'
print(a.upper())
print(A.lower())
print(t.title())
print(t.capitalize())
```

Методы проверок:

(возвращают bool)

1. s.isalpha() - проверяет, что все символы в строке - буквенные
2. s.isdigit() - все символы в строке - цифры
3. s.isalnum() - все символы в строке - цифры или буквы
4. s.isspace() - все символы в строке - пробельные
4. s.istitle() - строка является заголовком (результатом работы метода s.title())


```python
print('qwerty'.isalpha())
print('12345'.isdigit())
print('1234qwe'.isalnum())
print(' \t\n\xa0'.isspace()) # \t - табуляция, \xa0 - неразрывный пробел
print('The Lord Of The Rings'.istitle())
```

Метод замены части строки replace - возвращает новую строчку

s.replace(<что заменить>, <на что заменить>, <сколько раз заменить>)

(Сколько раз заменить - необязательный аргумент)

```python
s = 'qweqweqweqwe'
print(s.replace('q', '!', 2))
print(s.replace('q', '!'))
print('Строка осталась прежней:', s)
```

Метод strip - возвращает новую строчку

1. s.strip() - отгрызает все пробельные символы слева и справа
2. s.strip('123') - отгрызет все 1, 2 и 3 в любых сочетаниях слева и справа
3. s.lstrip() - отгрызет только слева
4. s.rstrip() - отгрызет только справа
