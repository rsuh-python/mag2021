```python
import matplotlib.pyplot as plt
```

```python
X = ['C', 'Python', 'C++', 'Java']
Y = [345, 123, 678, 345]

plt.plot(sorted(X), Y, '^g')  # два списка
plt.show()
```

```python
#Figure & Axes
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(X, Y, ':r')
plt.title('Languages of programming')
plt.xlabel('Names')
plt.ylabel('Number of users')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(X, Y, ':r')
ax.set_title('Languages', color='blue', size=20)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(X, Y)
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
ax.bar(X, Y)
plt.show()
```

```python

fig = plt.figure()
ax = fig.add_subplot(111)
ax.hist(Y, bins=4)
plt.show()
```

```python
plt.style.use(['fivethirtyeight'])
fig = plt.figure()
ax = fig.add_subplot(111)
ax.pie(Y, labels=X, autopct='%1.1f%%')
plt.show()
```

```python
Y1 = [123, 456, 345, 78]
Y2 = [34, 67, 890, 345]
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(X, Y1)
ax.plot(X, Y2)
plt.show()
```

```python
fig = plt.figure(figsize=(12, 9))
ax1 = fig.add_subplot(3, 1, 1)
ax1.plot(X, Y1)
ax1.set(title='Graph 1')
ax2 = fig.add_subplot(3, 2, 3)
ax2.set_title('Graph 2', size=10)
ax2.plot(X, Y2)
ax3 = fig.add_subplot(3, 3, 7)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
for ax in axes.flat:
    ax.plot(X, Y)
plt.show()
```

```python
from random import randrange
def myfunc(x, n):
    x = [randrange(0, x + 1) for i in range(n)]
    y = [i ** 2 for i in x]
    sns.barplot(x=x, y=y, color='lime', ci=None)
    plt.show()
```

```python
from ipywidgets import IntProgress
from IPython.display import display
import time

max_count = 100

f = IntProgress(min=0, max=max_count)
display(f)

for i in range(1, 101):
    f.value += 1
    time.sleep(0.1)
```

```python
from ipywidgets import interact, IntSlider

interact(myfunc, x=IntSlider(min=1, max=30, step=1, value=1), n=list(range(1, 11)))
```
